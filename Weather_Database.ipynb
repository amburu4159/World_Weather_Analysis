```python
# Import dependencies
import pandas as pd
import numpy as np
from citipy import citipy
from config import api_key
import requests
```

```python
# Create a new set of 2,000 random latitudes and longitudes.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Get the nearest city using the citipy module.
cities = list()
for coord in lat_lngs:
    cities_name = citipy.nearest_city(coord[0],coord[1]).city_name
    if cities_name not in cities:
        cities.append(cities_name)
print(cities[:15])
len(cities)
```

```python
# Perform an API call with the OpenWeatherMap.
units = "imperial"
weather_url ="https://api.openweathermap.org/data/2.5/weather?" 
query_url = f"{weather_url}appid={api_key}&units={units}&q="
response = requests.get(query_url+"kansas city").json() 
display(response)
```

```python
#set up list to hold response info
weather = []
counter = 0
#loop through the list of cities and perform a request for weather data for each city
for city in cities:
    #gets response for each city
    response = requests.get(query_url + city).json()
    #try/except statement to handle errors when city is not found on openweather site
    try:
        #create dictionary
        city_dict = {
            "City": response['name'],
            "Lat": response['coord']['lat'],
            "Lng": response['coord']['lon'],
            "Max Temp": response['main']['temp_max'],
            "Humidity": response['main']['humidity'],
            "Cloudiness": response['clouds']['all'],
            "Wind Speed": response['wind']['speed'],
            "Country": response['sys']['country'],
            "Current":response['weather'][0]['description']
        }
        weather.append(city_dict)
        counter = counter + 1
        print(f"{counter} City Name: {response['name']} / ID:{response['id']}")
    except:
        print("Missing city...skipping | " + city)
#city_weather
```

```python
# add data to a new data frame
cities_weather_df = pd.DataFrame(weather)
#Export Data to a csv file
cities_weather_df.to_csv("cities_weather")
cities_weather_df.head()
```
